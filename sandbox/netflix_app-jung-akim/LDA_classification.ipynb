```python
def one_string(texts, boolean_mask):
    thing = ' '.join([' '.join(item) for item in [row for row, boolean in zip(texts, boolean_mask) if boolean]])
    return re.sub(r'[ ]{2,}', ' ', thing)
```

```python
def create_wordcloud(img_directory, id, contour_width=3, contour_color='green', background_color="white"):
    topic_ids = np.array(LDA_preds)
    mask = np.array(Image.open(img_directory))
    stopwords = set(STOPWORDS)
    stopwords |= set(["movie", "thing", "love"])
    wc = WordCloud(background_color=background_color, max_words=15, mask=mask, stopwords=stopwords,
                   max_font_size=100, random_state=40, contour_width=contour_width, contour_color=contour_color)
    wc.generate(one_string(data_lemmatized, topic_ids == id))
    return wc
```

### Visualization of representative word embeddings with and UMAP

```python
term_doc_new = nlp_pipe.transform(reviews)
```

Create document-topic matrix

```python
doc_topic_matrix = np.array(pd.DataFrame([[prob for topic, prob in model.get_document_topics(item)] for item in term_doc_new]))
doc_topic_matrix[np.isnan(doc_topic_matrix)] = 0 # Zero out nan for TSNE
# np.argwhere(np.isnan(doc_topic_matrix))
```

Create term-topic matrix

```python
num_terms, num_topics = model.num_terms, model.num_topics

term_topic_matrix = np.array(pd.DataFrame([[prob for topic, prob in model.get_term_topics(item)] for item in range(num_terms)]))
term_topics = [model.get_term_topics(item) for item in range(num_terms)]
mat = np.array(sum([[(term, topic, prob) for (topic, prob) in topics] for term, topics in enumerate(term_topics) if len(topics) > 0], []))

topic = mat[:, 1].astype('int32')
term = mat[:, 0].astype('int32')
prob = mat[:, 2].astype('float32')

topic_term_matrix = coo_matrix((prob, (topic, term)), shape = (num_topics, num_terms)).toarray()
```

Create document-term matrix by multiplying the two matrices

```python
doc_term_matrix = np.matmul(doc_topic_matrix, topic_term_matrix)
```

Using the document-term matrix that has word embeddings, use TSNE

```python
LDA = pd.concat([pd.Series(LDA_preds), pd.Series([np.max(prob) for prob in LDA_proba])],axis=1)
LDA.columns = ['label', 'proba']

representative_docs, representative_docs_tsne, tokens_per_doc = [], [], []
representative_docs_umap = []
tsne = TSNE(n_components=2, random_state = 1)

for i in range(num_topics):
    LDA_ = LDA.loc[LDA.label==i,:]
    representative_docs.append(LDA_.index[np.argmax(LDA_.proba)])# Most representative documents per topic
    word_ids = term_doc_new[representative_docs[i]]
    embeddings, tokens = [], []
    for word_id, _ in word_ids:
        embeddings.append(doc_term_matrix[:,word_id])
        tokens.append(dictionary[word_id])
    tokens_per_doc.append(tokens)
#     representative_docs_tsne.append(tsne.fit_transform(np.array(embeddings)))
    mapper2d = umap.UMAP(n_components=2, n_neighbors=2, random_state=1).fit(np.array(embeddings))
    representative_docs_umap.append(mapper2d.embedding_)
```

```python
topic_dict = {0.: "Platform/Device", 1.: "User Experience", 2.: "Value", 3.: "Service", 4.: "Trouble-shooting", 5.:"Shows"}

ax = plt.figure(figsize=(10,10))
markers = ['o', 'x', '^', '+', 'v', 's']
colors = ['green', 'lightpink', 'gray', 'dodgerblue', 'darkred', 'goldenrod']
for label in range(num_topics):
    word_embeddings = representative_docs_umap[label]
    tokens = tokens_per_doc[label]
    x, y = word_embeddings[:, 0], word_embeddings[:, 1]
    plt.scatter(x, y, c=colors[label], label=topic_dict[label], alpha=0.5, marker = 'o', s=200);
    texts = [plt.text(x[i], y[i], tokens[i], ha='center', va='center') for i in range(len(x))]
    adjust_text(texts, avoid_points = False, expand_text = (1.0,1.5))
    
ax.legend(loc=6)
ax.savefig('images/{}'.format("lda_cluster"), dpi=200, transparent=True, bbox_inches = "tight");
```
