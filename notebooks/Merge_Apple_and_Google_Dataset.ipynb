# Merge Apple and Google

## Bolt

```python
apple = pd.read_csv(config['csv_input_local']['bolt_apple'],
                    usecols = ['userName','review','rating','date'])
apple = apple.reset_index(drop=True)
total_reviews = len(apple)

print(f'Total Apple reviews: {total_reviews} \n')
apple
```

```python
google = pd.read_csv(config['csv_input_local']['bolt_google'],
                     usecols = ['userName','content','score','at'])
google = google.rename(columns={'content': 'review','score': 'rating', 'at':'date'})

google = google.reset_index(drop=True)
total_reviews = len(google)

print(f'Total Google reviews: {total_reviews} \n')
google
```

```python
# merge both apple and google
df_merged = google.append(apple, ignore_index=True)
total_reviews = len(df_merged)

print(f'Total merged reviews: {total_reviews} \n')
df_merged
```

```python
df_merged.to_csv(config['csv_input_local']['bolt_apple_google'])
```
